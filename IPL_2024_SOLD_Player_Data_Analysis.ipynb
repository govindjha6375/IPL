# IPL 2024 Data Analysis: Auction Insights and Player Performance

## Overview:
The project aims to conduct a comprehensive analysis of the Indian Premier League (IPL) 2024 auction data, focusing on evaluating player prices, top sold players, and their on-field performance. Through data exploration, visualizations, and statistical analysis, this project seeks to uncover insights into team strategies, market trends, and the correlation between player valuation and performance.

## Objectives:
Analyze the distribution of player prices and identify the top sold players in the IPL 2024 auction.
Explore player performance metrics (batting average, strike rate, bowling economy, wickets taken, etc.) and assess the correlation between auction prices and on-field performance.
Investigate team-wise spending patterns and strategies adopted during the auction.
Identify market trends and factors influencing player valuation in the IPL 2024 auction.
## Tools and Techniques:
Libraries: Pandas, Matplotlib, Seaborn for data handling, analysis, and visualization.
Exploratory Data Analysis (EDA), Statistical Analysis, Data Visualization.
